````python
source_image = image = cv2.imread('cln1.bmp',0)

kernel = np.array([
    [0.,1.,0.],
    [1.,-4,1.],
    [0.,1.,0.]
]).reshape((3,3))

print(kernel)
````

````python
# Import SciPy library! used for convolutions!
import scipy.signal


# Performs convolution of source_image with kernel
output_image = scipy.signal.convolve2d(source_image, kernel)

# Filter images by assigning '0' to each pixel '< 0'
output_image[output_image < 0] = 0

# Filter images by assigning '255' to each pixel '> 255'
output_image[output_image > 255] = 255

# Display the output image
matplotlib.pyplot.imshow(output_image, cmap='gray')
matplotlib.pyplot.grid(False)
````
